```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
resources = pd.read_csv('resources.csv')
```

```python
train.head()
```

```python
test.head()
```

```python
resources.head()
```

```python
print('Shape of train->', train.shape, 'Shape of test->', test.shape, 'Shape of resources->', resources.shape)
```

```python
resources['total_cost'] = resources['quantity'] * resources['price']
```

```python
resources.head()
```

```python
resources_cost = resources.groupby(by = 'id', sort = False)['total_cost'].sum().reset_index()
```

```python
resources_cost
```

```python
train_resources = pd.merge(train, resources_cost, on = 'id')
test_resources = pd.merge(test, resources_cost , on = 'id')
```

```python
train_resources.head()
```

```python
test_resources.head()
```

```python
del train
del test
del resources
```

```python
train_resources.describe()
```

```python
y_train = train_resources['project_is_approved']
train_resources.drop(columns = ['project_is_approved'], inplace = True)
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import LabelEncoder

encode = LabelEncoder()
categorical_columns = ['school_state', 'project_grade_category']
for label in categorical_columns:
    train_resources[label] = encode.fit_transform(train_resources[label])
    test_resources[label] = encode.fit_transform(test_resources[label])
```

```python
train_resources.isnull().sum()
```

```python
train_resources.fillna(value= '', inplace = True)
test_resources.fillna(value = '', inplace = True)
```

```python
train_resources['project_essays'] = train_resources[['project_essay_1', 'project_essay_2', 'project_essay_3', 'project_essay_4']].apply(lambda x: ' '.join(x), axis = 1)
test_resources['project_essays'] = test_resources[['project_essay_1', 'project_essay_2', 'project_essay_3', 'project_essay_4']].apply(lambda x: ' '.join(x), axis = 1)
```

```python
train_resources['project_subjects'] = train_resources[['project_subject_categories', 'project_subject_subcategories']].apply(lambda x: ' '.join(x), axis = 1)
test_resources['project_subjects'] = test_resources[['project_subject_categories', 'project_subject_subcategories']].apply(lambda x: ' '.join(x), axis = 1)
```

```python
encode = LabelEncoder()

train_resources['project_subjects'] = encode.fit_transform(train_resources['project_subjects'])
test_resources['project_subjects'] = encode.fit_transform(test_resources['project_subjects'])
```

```python
train_resources['teacher_prefix'] = encode.fit_transform(train_resources['teacher_prefix'])
test_resources['teacher_prefix'] = encode.fit_transform(test_resources['teacher_prefix'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(ngram_range = (1, 2))
train_project_essays = tf.fit_transform(train_resources['project_essays'])
test_project_essays = tf.fit_transform(test_resources['project_essays'])
train_project_resource = tf.fit_transform(train_resources['project_resource_summary'])
test_project_resource = tf.fit_transform(test_resources['project_resource_summary'])
```

```python
train_resources['project_submitted_datetime'] = pd.to_datetime(train_resources['project_submitted_datetime'])
test_resources['project_submitted_datetime'] = pd.to_datetime(test_resources['project_submitted_datetime'])
```

```python
train_resources['date'] = train_resources['project_submitted_datetime'].dt.date
train_resources['year'] = train_resources['project_submitted_datetime'].dt.year
train_resources['month'] = train_resources['project_submitted_datetime'].dt.month
train_resources['hours'] = train_resources['project_submitted_datetime'].dt.hour
train_resources['minutes'] = train_resources['project_submitted_datetime'].dt.minute
train_resources['seconds'] = train_resources['project_submitted_datetime'].dt.second

test_resources['date'] = test_resources['project_submitted_datetime'].dt.date
test_resources['year'] = test_resources['project_submitted_datetime'].dt.year
test_resources['month'] = test_resources['project_submitted_datetime'].dt.month
test_resources['hours'] = test_resources['project_submitted_datetime'].dt.hour
test_resources['minutes'] = test_resources['project_submitted_datetime'].dt.minute
test_resources['seconds'] = test_resources['project_submitted_datetime'].dt.second
```

```python
final_columns = ['school_state','teacher_prefix', 'year', 'month', 'hours', 'minutes', 'seconds', 'project_grade_category', 'total_cost']
```

```python
from scipy.sparse import hstack, csr_matrix

X_train_full = csr_matrix(hstack([train_resources[final_columns].values,train_project_essays,train_project_resource]))
X_test_full = csr_matrix(hstack([test_resources[final_columns].values,test_project_essays,test_project_resource]))
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

clf_1 = MultinomialNB()
clf_2 = SVC()
clf_3 = RandomForestClassifier()
clf_4 = LogisticRegression()

print('Classifier 3: Random Forest')
print('=============Training================')
clf_3 = clf_3.fit(X_train_full, y_train)
print('=============Predicting==============')
y_test_3 = clf_3.predict_proba(X_test_full)[:,1]
print(y_test_3)
output = pd.concat([pd.DataFrame({'id':test_resources['id']}), pd.DataFrame(y_test_3, columns = ['project_is_approved'])], axis =1)
output.to_csv('random_forest_submission.csv', index = False)
print('Accuracy of Random Forest after submission on Kaggle = ')
```
