```python
drop_indices = []
for index in cat_series.index:
    if cat_series[index] == True:
        drop_indices.append(index)
```

```python
drop_indices = list(drop_indices)
print(drop_indices)
train = train.drop(columns = drop_indices, axis = 1)
```

```python
#train.dtypes
category_columns = list(train.select_dtypes(['object']).columns)
print(category_columns)
for column in category_columns:
    train[column] = train[column].astype('category')
```

```python
train.info()
```

```python
train.describe()
```

```python
float_columns = list(train.select_dtypes(['float','int']).columns)
print(float_columns)
```

```python
#check the standard deviation of all numeirc columns
train[float_columns].std()
```

```python
# proportion of missing entries in each float columns
train[float_columns].isnull().sum() / train.shape[0]
```

```python
# filling missing values
float_mean = train[float_columns].mean()
for column in float_columns:
    missing_rows = train[column].isnull()
    if missing_rows.sum() > 0:
        train.ix[missing_rows.tolist(), column] = float_mean[column]
```

```python
# all null values filled
train[float_columns].isnull().sum() / train.shape[0]
```

```python
# Getting count of categories in each categorical column
categories_number_cols = train[category_columns].apply(lambda col: len(col.cat.categories))
print(categories_number_cols)
```

```python
# Neglecting all with more than 500 categories
collapsed_categories = category_columns
category_cols = categories_number_cols[categories_number_cols <= 500].index
category_cols
```

```python
# Handling categorical data
for col in category_cols:
    missing_rows = train[col].isnull()
    
    if missing_rows.sum() > 0:
        train[col].cat.add_categories('MISSING', inplace = True)
        train.ix[missing_rows.tolist(), col] = 'MISSING'
    
    temp = train[col].copy()
    
    for category in temp.cat.categories:
        matching_cat_rows = temp == category
        
        if matching_cat_rows.sum() < 0.05*train.shape[0]:
            if 'OTHER' not in train[col].cat.categories:
                train[col].cat.add_categories('OTHER', inplace = True)
            train.ix[matching_cat_rows, col] = 'OTHER'
            train[col].cat.remove_categories(category, inplace = True)
```

```python
categories_number_cols = train[category_columns].apply(lambda col: len(col.cat.categories))
print(categories_number_cols)
```

```python
remove_columns = list(set(category_columns) - set(category_cols))
remove_columns
```

```python
train.drop(columns = remove_columns, axis = 1, inplace = True)
```

```python
train.head()
```

```python
category_columns = list(train.select_dtypes(['category']).columns)
categories_number_cols = train[category_columns].apply(lambda col: len(col.cat.categories))
print(categories_number_cols)
```

```python
remove_cat_columns = []

for column in list(category_columns):
    if len(train[column].cat.categories) == 1 or (len(set(train[column].cat.categories) -set(['MISSING', 'OTHER'])) < 2):
        remove_cat_columns.append(column)
remove_cat_columns
```

```python
train.drop(columns = remove_cat_columns, inplace = True, axis = 1)
```

## Validation split

```python
from random import seed
from sklearn.cross_validation import StratifiedShuffleSplit

RANDOM_SEED = 24
seed(RANDOM_SEED)

test_proportion = .6
split = StratifiedShuffleSplit(
    y=appetency,
    n_iter=1,
    test_size=test_proportion,
    random_state=RANDOM_SEED)

for train_valid_indices, test_indices in split:
    X_train_valid, X_test = train.iloc[train_valid_indices, :], train.iloc[test_indices, :]
    appetency_train_valid, appetency_test = appetency.iloc[train_valid_indices], appetency.iloc[test_indices]
    
    
valid_proportion_of_train_valid = .25
split = StratifiedShuffleSplit(
    y=appetency_train_valid,
    n_iter=1,
    test_size=valid_proportion_of_train_valid,
    random_state=RANDOM_SEED)

for train_indices, valid_indices in split:
    X_train, X_valid = X_train_valid.iloc[train_indices, :], X_train_valid.iloc[valid_indices, :]
    appetency_train, appetency_valid = appetency_train_valid.iloc[train_indices], appetency_train_valid.iloc[valid_indices]
```

```python
appetency_train
```

# Classification Model

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(
    n_estimators = 1000,
    criterion = 'entropy',
    min_samples_split = 100,
    max_features = 'log2',
    random_state = 42,
    n_jobs = -1
)

clf.fit(X_train, appetency_train)
```

```python
train_predicted = clf.predict(X_valid)
```

```python
from sklearn.metrics import accuracy_score

print('Accuracy on validation data = ', accuracy_score(appetency_valid, train_predicted))
```
