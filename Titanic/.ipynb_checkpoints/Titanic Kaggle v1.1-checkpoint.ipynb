# Feature Engineering

```python
# Generally we combine test and train data to do feature engineering so that we consider all features (i.e sometimes features are different in train and test set)
```

```python
test = pd.read_csv('test.csv')
test
```

```python
# We defined a function here because reusability is important in data science.
def getCombinedData():
    train = pd.read_csv('train.csv')
    test = pd.read_csv('test.csv')
    targets = train['Survived']
    train.drop('Survived', axis= 1, inplace=True)
    
    combined = train.append(test)
    combined.reset_index(inplace=True)
    combined.drop('index', axis= 1, inplace=True)
    
    return combined
```

```python
combined = getCombinedData()
```

```python
combined.shape # train + test instances = 1309  
```

```python
# What to do with Person names?
# Let us extract person title...
# What are the different person titles present in the dataset?
titles = set(combined['Name'].map(lambda x: x.split(',')[1].split('.')[0].strip()))
```

```python
titles
```

```python
'''
Capt, Col, Major, Rev, Dr --> Officer
Jonkheer, Lady, Sir, Don, Dona, the Countess  --> Nobility
Mr -->Mr
Miss,Mlle(English equivalent to Miss) -->Miss
Master -->Master
Mrs, Mme(Madam), Ms -->Mrs
'''
def getTitles():
    global combined
    
    combined['Title'] = combined['Name'].map(lambda x: x.split(',')[1].split('.')[0].strip())
    
    Title_dict = {
        'Capt': 'Officer',
        'Col': 'Officer',
        'Major': 'Officer',
        'Rev': 'Officer',
        'Dr': 'Officer',
        'Jonkheer': 'Nobility',
        'Lady': 'Nobility',
        'Sir': 'Nobility',
        'Don': 'Nobility',
        'Dona': 'Nobility',
        'the Countess': 'Nobility',
        'Mr': 'Mr',
        'Miss': 'Miss',
        'Mlle': 'Miss',
        'Master': 'Master',
        'Mrs': 'Mrs',
        'Mme': 'Mrs',
        'Ms': 'Mrs'
    }
    
    combined['Title'] = combined['Title'].map(Title_dict)
```

```python
# Previously we replaced missing Age with the combined median of Age but that can go wrong sometimes. For example see this

grouped_train = combined.head(891).groupby(['Sex','Pclass','Title'])
grouped_train_median = grouped_train.median()

grouped_test = combined.iloc[891:].groupby(['Sex','Pclass','Title'])
grouped_test_median = grouped_test.median()
```

```python
grouped_train_median
```

```python
 # Above we see that female with pclass1 Miss Title has median age = 30 and with Mrs title has median Age 40 and similarly we can see several such cases
```

```python
# process missing ages

def fill_missing_ages(row, grouped_median):
    
    if row['Sex'] == 'female' and row['Pclass'] == 1:
        if row['Title'] == 'Miss':
            return grouped_median.loc['female', 1, 'Miss']['Age']
        elif row['Title'] == 'Mrs':
            return grouped_median.loc['female', 1, 'Mrs']['Age']
        elif row['Title'] == 'Nobility':
            return grouped_median.loc['female', 1, 'Nobility']['Age']
        elif row['Title'] == 'Officer':
            return grouped_median.loc['female', 1, 'Officer']['Age']
        
    elif row['Sex'] == 'female' and row['Pclass'] == 2:
        if row['Title'] == 'Miss':
            return grouped_median.loc['female', 2, 'Miss']['Age']
        elif row['Title'] == 'Mrs':
            return grouped_median.loc['female', 2, 'Mrs']['Age']
        
    elif row['Sex'] == 'female' and row['Pclass'] == 3:
        if row['Title'] == 'Miss':
            return grouped_median.loc['female', 3, 'Miss']['Age']
        elif row['Title'] == 'Mrs':
            return grouped_median.loc['female', 3, 'Mrs']['Age']
        
    if row['Sex'] == 'male' and row['Pclass'] == 1:
        if row['Title'] == 'Master':
            return grouped_median.loc['male', 1, 'Master']['Age']
        elif row['Title'] == 'Mr':
            return grouped_median.loc['male', 1, 'Mr']['Age']
        elif row['Title'] == 'Nobility':
            return grouped_median.loc['male', 1, 'Nobility']['Age']
        elif row['Title'] == 'Officer':
            return grouped_median.loc['male', 1, 'Officer']['Age']
        
    elif row['Sex'] == 'male' and row['Pclass'] == 2:
        if row['Title'] == 'Master':
            return grouped_median.loc['male', 2, 'Master']['Age']
        elif row['Title'] == 'Mr':
            return grouped_median.loc['male', 2, 'Mr']['Age']
        elif row['Title'] == 'Officer':
            return grouped_median.loc['male', 2, 'Officer']['Age']
        
    elif row['Sex'] == 'male' and row['Pclass'] == 3:
        if row['Title'] == 'Master':
            return grouped_median.loc['male', 3, 'Master']['Age']
        elif row['Title'] == 'Mr':
            return grouped_median.loc['male', 3, 'Mr']['Age']
        

combined.head(891).Age = combined.head(891).apply(lambda r: fill_missing_ages(r,grouped_train_median) if np.isnan(r['Age'])
                                                 else r['Age'], axis=1)
# axis = 1 as we want to apply to each row.

combined.iloc[891:].Age = combined.iloc[891:].apply(lambda r: fill_missing_ages(r,grouped_test_median) if np.isnan(r['Age'])
                                                 else r['Age'], axis=1)

    
        
```

```python
# First let us get rid of the 'Name' column as we have no information left to extract from it
def clean_names():
    global combined
    
    combined.drop('Name', inplace=True, axis=1)
    
    # Now use one hot encoding or pandas get_dummies() method to encode categorical data
    titles_dummies = pd.get_dummies(combined['Title'], prefix='Title')
    combined = pd.concat([combined,titles_dummies], axis=1)
    combined.drop('Title',axis=1, inplace=True)
```

```python
# Process missing fare value
def fill_missing_fares():
    global combined
    
    combined.head(891)['Fare'].fillna(combined.head(891)['Fare'].mean(), inplace=True)
    combined.iloc[891:]['Fare'].fillna(combined.iloc[891:]['Fare'].mean(), inplace=True)
    
fill_missing_fares()
```

```python
combined.info()
```

```python
# handle missing Embarked entries and creating dummy variables for the categorical data
def handle_embarked():
    global combined
    
    combined.Embarked.fillna('S', inplace=True)
    
    embarked_dummies = pd.get_dummies(combined['Embarked'], prefix='Embarked')
    
    combined = pd.concat([combined,embarked_dummies], axis=1)
    combined.drop('Embarked', axis=1, inplace=True)
    
```

```python
handle_embarked()
```

```python
combined.info()
```

```python
# Processing cabin
set(combined['Cabin'].map(lambda x: str(x)[0])) #NaN is converted as null by str so 'n'
```

```python
# replace missing cabin details by 'U' and encode categorical data
def handle_cabin():
    global combined
    
    combined['Cabin'].fillna('U', inplace=True)
    
    combined['Cabin'] = combined['Cabin'].map(lambda x: x[0])
    
    # dummies
    cabin_dummies = pd.get_dummies(combined['Cabin'], prefix='Cabin')
    combined = pd.concat([combined,cabin_dummies], axis= 1)
    
    combined.drop('Cabin', axis=1, inplace=True)
```

```python
handle_cabin()
```

```python
combined.info()
```

```python
# process sex
def process_sex():
    global combined
    
    combined['Sex'] = combined['Sex'].map({'male':1,'female':0})
```

```python
process_sex()
```

```python
combined.info()
```

```python
# Process Pclass
def process_pclass():
    global combined
    
    pclass_dummies = pd.get_dummies(combined['Pclass'], prefix='Pclass')
    combined = pd.concat([combined,pclass_dummies], axis=1)
    
    combined.drop('Pclass', axis=1, inplace=True)
```

```python
process_pclass()
```

```python
combined.head()
```

```python
def handle_ticket():
    
    global combined
    
    def cleanTicket(ticket):
        ticket = ticket.replace('.','')
        ticket = ticket.replace('/','')
        ticket = ticket.split()
        ticket = map(lambda t : t.strip(), ticket)
        ticket = list(filter(lambda t : not t.isdigit(), ticket))
        if len(ticket) > 0:
            return ticket[0]
        else: 
            return 'UNKN'
    


    combined['Ticket'] = combined['Ticket'].map(cleanTicket)
    tickets_dummies = pd.get_dummies(combined['Ticket'], prefix='Ticket')
    combined = pd.concat([combined, tickets_dummies], axis=1)
    combined.drop('Ticket', inplace=True, axis=1)
```

```python
handle_ticket()
```

```python
# Family 
# Family size, assumption large family more likely to be saved

def process_family():
    
    global combined
    combined['FamilySize'] = combined['Parch'] + combined['SibSp'] + 1
    
    combined['Singleton'] = combined['FamilySize'].map(lambda s: 1 if s == 1 else 0)
    combined['SmallFamily'] = combined['FamilySize'].map(lambda s: 1 if 2<=s<=4 else 0)
    combined['LargeFamily'] = combined['FamilySize'].map(lambda s: 1 if 5<=s else 0)
```

```python
process_family()
```
