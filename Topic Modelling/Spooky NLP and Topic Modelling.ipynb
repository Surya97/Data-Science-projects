```python
from wordcloud import WordCloud, STOPWORDS
from scipy.misc import imread
import matplotlib.pyplot as plt
```

```python
# for EAP
eap_mask = imread('./eap.png', flatten = True)
plt.figure(figsize = (16,13))

wc = WordCloud(background_color = 'black', max_words = 15000, mask = eap_mask, stopwords = STOPWORDS, max_font_size = 40)
wc.generate(' '.join(eap))
plt.title('Edgar Allen Poe', fontsize = 20)
plt.imshow(wc.recolor(colormap = 'PuBu', random_state = 17), alpha = 0.98)
plt.axis('off')
plt.show()
```
