```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from gensim.models import word2vec
import re
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
import nltk
```

```python
def comment_to_wordlist(comment, remove_stopwords = False):
    #Remove HTML tags
    comment_text = BeautifulSoup(comment).get_text()
    #Remove all non-letters
    comment_text = re.sub("[^a-zA-Z]"," ", comment_text)
    #Convert to lower case and split into words
    words = comment_text.lower().split()
    
    #remove stopwords if specified
    if remove_stopwords:
        stops = set(stopwords.words("english"))
        words = [w for w in words if not w in stops]
    return words

tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
def comment_to_sentences(comment, tokenizer, remove_stopwords = False):
    raw_sentences = tokenizer.tokenize(comment.strip())
    sentences = []
    for raw_sentence in raw_sentences:
        if len(raw_sentence) > 0:
            sentences.append(comment_to_wordlist(raw_sentence, remove_stopwords))
    
    return sentences

sentences = []
print('Parsing sentences from training set')
for comment in train['comment_text']:
    sentences+=comment_to_sentences(comment, tokenizer)
```

```python
model = word2vec.Word2Vec(sentences, size = 300, window = 10, min_count = 40, workers = 4)
```

```python
def create_bag_of_centroids( wordlist, word_centroid_map ):
    #
    # The number of clusters is equal to the highest cluster index
    # in the word / centroid map
    num_centroids = max( word_centroid_map.values() ) + 1
    #
    # Pre-allocate the bag of centroids vector (for speed)
    bag_of_centroids = np.zeros( num_centroids, dtype="float32" )
    #
    # Loop over the words in the review. If the word is in the vocabulary,
    # find which cluster it belongs to, and increment that cluster count 
    # by one
    for word in wordlist:
        if word in word_centroid_map:
            index = word_centroid_map[word]
            bag_of_centroids[index] += 1
    #
    # Return the "bag of centroids"
    return bag_of_centroids
```

```python
from sklearn.cluster import KMeans
num_clusters = model.wv.syn0.shape[0] // 5

kmeans_clustering = KMeans( n_clusters = num_clusters )
idx = kmeans_clustering.fit_predict( model.wv.syn0 )

word_centroid_map = dict(zip( model.wv.index2word, idx ))
```

```python
clean_train_comments = []
for comment in train['comment_text']:
    clean_train_comments.append(comment_to_wordlist(comment, remove_stopwords = True))
```

```python
clean_test_comments = []
for comment in test['comment_text']:
    clean_test_comments.append(comment_to_wordlist(comment, remove_stopwords = True))
```

```python
# Pre-allocate an array for the training set bags of centroids (for speed)
print(num_clusters)
train_centroids = np.zeros( (train['comment_text'].size, num_clusters), dtype="float32" )

# Transform the training set comments into bags of centroids
counter = 0
for comment in clean_train_comments:
    train_centroids[counter] = create_bag_of_centroids( comment, word_centroid_map )
    counter += 1

# Repeat for test comments 
test_centroids = np.zeros(( test['comment_text'].size, num_clusters), dtype="float32" )

counter = 0
for comment in clean_test_comments:
    test_centroids[counter] = create_bag_of_centroids( comment, word_centroid_map )
    counter += 1
len(test_centroids)
```

```python
temp = train[train.columns[2:]]
target = [0]*(temp.shape[0])
counter = 0
for ind,row in temp.iterrows():
    if row['toxic'] == 1:
        target[counter] = 1
    elif row['severe_toxic'] == 1:
        target[counter] = 2
    elif row['obscene'] == 1:
        target[counter] = 3
    elif row['threat'] == 1:
        target[counter] = 4
    elif row['insult'] == 1:
        target[counter] = 5
    elif row['identity_hate'] == 1:
        target[counter] = 6

    counter+=1

target[:10]
```

```python
from sklearn.naive_bayes import MultinomialNB
clf1 = MultinomialNB()
target = np.array(target)
clf1.fit(train_centroids, target)
```
